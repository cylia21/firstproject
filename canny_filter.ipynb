```
import cv2 as cv 
import matplotlib.pyplot as plt
```

```
### File Path 
cat_path = "/content/drive/My Drive/chat.jpg"
```

```
### Load image 
file = cv.imread(cat_path, 0)
```

```
### Image dimensions h x w 
file.shape
```

```
detector = cv.Canny(file, 100, 200)
```

```
### Show origin image
plt.imshow(file)
```

```
### Show edges 
plt.imshow(detector) 
```

```
plt.subplot(121), plt.imshow(file, cmap='gray')
plt.title("Image du chat -- origine")
plt.subplot(122),  plt.imshow(detector, cmap='gray')
plt.title("Filtre de Canny")
plt.show()
```
